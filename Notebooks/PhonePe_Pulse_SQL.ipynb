## Create & Show the tables

```python
#Create the tables
for table_name, df in dataframes.items():
    try:
        df.to_sql(name=table_name, con=engine, if_exists='replace', index=False)
        print(f"Table '{table_name}' created and data inserted successfully! ({len(df)} rows)")
    except Exception as e:
        print(f"Error creating table '{table_name}': {e}")
```

## Query

### Total transaction amount by year

```python
query = """
SELECT Year, SUM(Transaction_amount) AS Total_amount
FROM agg_trans
GROUP BY Year
ORDER BY Total_amount DESC;
"""
yearly_trans = pd.read_sql(query, engine)
print(f"Total transaction amount by year:")
display(yearly_trans)
```

### Total transaction by year & quarter

```python
query='''SELECT Year,Quarter,SUM(Transaction_count) AS Total_transaction
FROM agg_trans
GROUP BY Year,Quarter;'''

print(f"Total transaction by year & quarter:")
total_trans=pd.read_sql(query,engine)
display(total_trans)
```

### Transactions type distribution by Transactions amount

```python
query = """
SELECT Transaction_type, SUM(Transaction_amount) AS Total_amount
FROM agg_trans
GROUP BY Transaction_type
ORDER BY Total_amount ASC;
"""
type_trans = pd.read_sql(query, engine)
print(f"Transaction type distribution by total amount Of transaction:")
display(type_trans)
```

### Regional distribution of total transactions

```python
query='''
SELECT Region, SUM(Transaction_amount) AS Total_amount, SUM(Transaction_count) AS Total_transaction
FROM agg_trans
GROUP BY Region
ORDER BY Total_transaction DESC;
'''
region=pd.read_sql(query,engine)
print(f"Regional distribution of total transactions:")
display(region)
```

### Top 10 states by transaction amount

```python
query='''
SELECT State, SUM(Transaction_amount) AS Total_amount
FROM agg_trans
GROUP BY State
ORDER BY Total_amount DESC
LIMIT 10;'''

state=pd.read_sql(query,engine)
print(f"Top 10 states by transaction amount:")
display(state)
```

### Top brands by transaction count

```python
query='''
SELECT Brand, SUM(Transaction_count) AS Total_Transaction
FROM agg_user
GROUP BY Brand
ORDER BY Total_Transaction DESC
LIMIT 10;'''

brand=pd.read_sql(query,engine)
print(f"Top 10 states by transaction count:")
display(brand)
```

### Yearly growth in transaction amount, transaction count

```python
query='''SELECT Year, SUM(Transaction_amount) AS Total_Amount,SUM(Transaction_count) AS Total_Transaction
FROM agg_trans
GROUP BY Year
ORDER BY Year;'''

growth=pd.read_sql(query,engine)
print("Yearly growth in transaction amount, transaction count:")
display(growth)
```

### Yearly Growth of Registered Users

```python
query='''
SELECT Year, SUM(Registered_users) AS Total_users
FROM map_user
GROUP BY Year
ORDER BY Year;
'''

user=pd.read_sql(query,engine)
print(f"Yearly Growth of Registered Users:")
display(user)
```

### Registered Users by Region

```python
query='''
SELECT Region, SUM(Registered_users) AS Total_users
FROM map_user
GROUP BY Region
ORDER BY Region;
'''

user=pd.read_sql(query,engine)
print(f"Registered users by Region:")
display(user)
```

### Top Districts by Transaction Volume

```python
query='''SELECT District, SUM(Transaction_amount) AS total_amount
FROM top_trans_dist
GROUP BY District
ORDER BY total_amount DESC
LIMIT 10;
'''

dist=pd.read_sql(query,engine)
print("Top districts:")
display(dist)
```

### State-wise App Opens vs Registered Users

```python
query='''
SELECT 
  State, 
  SUM(Registered_users) AS total_users, 
  SUM(App_opens) AS total_app_opens,
  ROUND(SUM(App_opens) / SUM(Registered_users), 2) AS opens_per_user
FROM map_user
GROUP BY State
ORDER BY opens_per_user DESC
LIMIT 10;
'''

q=pd.read_sql(query,engine)
display(q)
```

### Compare Average Transaction Amount per State

```python
query='''
SELECT 
  State, 
  SUM(Transaction_amount)/SUM(Transaction_count) AS avg_transaction
FROM agg_trans
GROUP BY State
ORDER BY avg_transaction DESC
LIMIT 5; '''

avg=pd.read_sql(query,engine)
print(f"Average Transaction Amount")
display(avg)
```

### Top performing pincodes

```python
query='''SELECT Pincode, SUM(Transaction_amount) AS total_amount
FROM top_trans_pin
GROUP BY Pincode
ORDER BY total_amount DESC
LIMIT 10;
'''

q=pd.read_sql(query,engine)
print("Top performing pincodes:")
display(q)
```

### Correlation Insight â€“ Transaction Amount vs Registered Users by District

```python
query='''
SELECT u.District,
    SUM(u.Registered_users) AS Total_users,
    SUM(t.Transaction_amount) AS total_amount
FROM map_user u
INNER JOIN map_trans t
    ON u.District = t.District AND u.State = t.State AND u.Year = t.Year
GROUP BY u.District
ORDER BY total_amount DESC;
'''

q=pd.read_sql(query,engine)
print(f"Correlation Insight- Transaction Amount vs Registered Users by District:")
display(q)
```
